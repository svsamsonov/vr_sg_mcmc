```python
import numpy as np
import numpy.polynomial as P
import scipy as sp
from matplotlib import pyplot as plt
from tqdm import tqdm
#from sklearn.preprocessing import PolynomialFeatures
from multiprocessing import Pool
import multiprocessing
import ZVnbrosse
from potentials import PotentialOne, PotentialTwo, PotentialOnePrime
from zv_cv import Eval_ZVCV_Gaus
from samplers import MCMC_sampler,Generate_train,ULA_light
from baselines import set_function,construct_ESVM_kernel,GenerateSigma
from martingale import approx_q,test_traj
from optimize import Run_eval_test,optimize_parallel_new 
from utils import *
import copy
```

```python
N_burn = 1*10**4 # Burn in period
N_train = 1*10**4 # Number of samples on which we optimize
step = 0.05 # Step size
#put 0.5 for MALA
#step = 0.2
n_traj = 100 # Number of independent MCMC trajectories for test
f_type = "sum"
```

Choose density parameters

```python
d = 2
B = np.pi/2
M = 10.0
Cur_pot = PotentialOne(B,M,d)
#mu = 0.5
#sigma = 0.5
#M = 1.0
#Cur_pot = PotentialTwo(M,mu,sigma,d)
```

### Visulaize level set of the potential

```python
###
def U(x, y):
    return -1./(2*M)*x**2 - 1./2*(y-np.sin(B*x))**2

x = np.linspace(-5, 5, 200)
y = np.linspace(-5, 5, 200)

X, Y = np.meshgrid(x, y)
Z = np.exp(U(X, Y))
```

```python
plt.imshow(Z, extent=[-3, 3, -3, 3], origin='lower',
           cmap='RdGy')
plt.colorbar()
plt.axis(aspect='image')
```

### Generate data

```python
sampler = {"sampler":"RWM","burn_type":"full","main_type":"full"} # Sampling method

if sampler["sampler"] == "ULA":
    res = Generate_train(n_traj, sampler, Cur_pot, step, N_burn, N_train, d)
    res = np.asarray(res)
    traj,traj_grad = res[:,0,:,:],res[:,1,:,:]
else:
    res = Generate_train(n_traj, sampler, Cur_pot, step, N_burn, N_train, d)
    traj = []
    traj_grad = []
    for i in range(len(res)):
        traj.append(res[i][0])
        traj_grad.append(res[i][1])
        print("accepted = ",res[i][2])
    traj = np.asarray(traj)
    traj_grad = np.asarray(traj_grad)
```

```python
print(traj.shape)
print(traj_grad.shape)
traj_grad = (-1)*traj_grad
```

### Visualisation

```python
from statsmodels.tsa.stattools import acf
sample_acf = acf(traj[0,:,1], unbiased = True, nlags = 100, fft = True)

plt.figure(figsize=(20,5))
plt.title('Sample  ACF',fontsize=16)
plt.plot(sample_acf, label='sample acf')
plt.legend(fontsize=16)

plt.show()
```

### Variance reduction part

```python
bn = 100
W_test = construct_ESVM_kernel(N_train,bn)
```

```python
test_seed = 1453
f_type = "sum"
nbcores = multiprocessing.cpu_count()
trav = Pool(nbcores)
res = trav.starmap(Eval_ZVCV_Gaus, [(traj[i,:,:],traj_grad[i,:,:],f_type,None,W_test) for i in range (n_traj)])
trav.close()
```

```python
res_arr = np.asarray(res)
print(res_arr.shape)
```
